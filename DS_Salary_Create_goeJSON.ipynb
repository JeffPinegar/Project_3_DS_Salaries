### Read in the geoJSON data

```python
# Import data file
# My Jupyter notebook is in the current directory " . "
# There is a sub directory "Resources" and the file "IPC shipped from Germany by serial number 2022 (until 5.12.2022).csv" is located there.
geoData = os.path.join('.','data', 'countries.geojson')            

# Opens the CVS file and read it in as a panda dataframe then print the head.
# df_raw = pd.read_csv(salary_data, index_col=0, encoding="utf-8")
df_geo = gpd.read_file(geoData)                   
df_geo.head(10)
```

---
### Read in the average salary for each country by year

```python
# Import data file
# My Jupyter notebook is in the current directory " . "
# There is a sub directory "Resources" and the file "IPC shipped from Germany by serial number 2022 (until 5.12.2022).csv" is located there.
salary_by_country_data = os.path.join('.','data', 'Data_Science_Salary_by_Country.csv')            

# Opens the CVS file and read it in as a panda dataframe then print the head.
# df_raw = pd.read_csv(salary_data, index_col=0, encoding="utf-8")
df_cs = pd.read_csv(salary_by_country_data, encoding="utf-8", index_col=False)                   
df_cs.head()
```

---
### Merge the data frame and geodataframe

```python
df_final = df_geo.merge( df_cs, left_on='ISO_A3', right_on='cc_3_char')
df_final.head()
```

```python
geoData = os.path.join('.','data', 'cc_Salary.geojson')
df_final.to_file(geoData, driver='GeoJSON')
```
